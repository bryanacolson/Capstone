```python
from pandasql import sqldf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import io
import requests
from ast import literal_eval
import glob
```

```python
project_csv = pd.concat([pd.read_csv(x)for x in glob.glob('../csvs_4/*.csv')], ignore_index = True)
```

```python
project_csv = project_csv.dropna()
project_csv
```

```python
project_csv.columns = ['Explicit', 'Popularity', 'Artist', 'Song', 'Duration ms', 'uri']
```

```python
project_csv = project_csv.sort_values(by= 'Popularity', ascending=False)
project_csv
```

```python
project_csv = project_csv.explode(column = "Song")
project_csv = project_csv.dropna()
project_csv
```

```python
popular_song_1000 = project_csv.sort_values(by= 'Popularity', ascending = False).head(1000)
popular_song_1000
```

```python
popular_song_20 = project_csv.sort_values(by= 'Popularity', ascending = False).head(20)
popular_song_20
```

```python
popular_song_100 = project_csv['Song'].value_counts().head(100).index.to_list()
popular_song_100
```

```python
song_graph = popular_song_20.plot(figsize=(20,10),
                             x= 'Artist',
                             y= 'Popularity',
                             title = 'Artist Popularity',
                             xlabel = 'Artist',
                             ylabel = 'Popularity',
                             kind = 'bar',
                             color = 'cadetblue',
                             
                             alpha = 0.8)
for bars in song_graph.containers:
    song_graph.bar_label(bars)
plt.xticks(rotation = 90);
```
